```python
# @title Titre par défaut
# Dropping (delete) missing values
copie_data = data.dropna(inplace = False) # inplace = True : changes will be made inplace , inplace = false : you have to assign the result to a new Dataframe like this example
copie_data
```

```python
# Remplir avec la valeur moyenne
for col in numerical_features :
    data[col].fillna(data[col].mean() , inplace = True)


# Remplir avec la valeur médiane
'''for col in numerical_features :
    data[col].fillna(data[col].median() , inplace = True)'''

# Remplir avec la valeur la plus fréquente (mode, pour colonnes catégoriques)
for col in categorical_features :
    data[col].fillna(data[col].mode()[0] , inplace = True)

```

#Feature selection and Splitting Data 🎶✂️

```python
# Feature Extraction
x = data.iloc[ : , 1:7] # x (features)
y = data.iloc[ : , -1] # y (target)


from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)

print("X_train:\n", x_train)
print("X_test:\n", x_test)
print("Y_train:\n", y_train)
print("Y_test:\n", y_test)
```

#Model Selection And Training 🎶🤖

```python
#Standardization with StandardScaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

print(x_train)
print(x_test)
```

```python
#Classification with KNN

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

KNN = KNeighborsClassifier(n_neighbors = 5)
KNN.fit(x_train,y_train)
y_predict = KNN.predict(x_test)

print("Accuracy = " , accuracy_score(y_predict,y_test))
```

```python
#Classification with SVM

from sklearn.svm import SVC

classifier = SVC()
classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_test)
print("Accuracy = ", accuracy_score(y_pred,y_test)) #Evaluation
```

#Test the model 🎶📝📈
